## Standardization & Transformation

#### 1. Standard Scaler(Standardization)
We try to bring all the variables/features to similar scale.
STandardization means centering the variable to zero. Z = (X- X mean)/Std

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic_train.csv', usecols=['Pclass','Age','Fare','Survived'])
df.tail()
```

```python
df['Age'].fillna(df.Age.median(),inplace=True)
df.isna().sum()
```

```python
df.Age.median()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaler = scaler.fit_transform(df)
df_scaler
```

```python
type(df_scaler)
```

```python
df_1=pd.DataFrame(df_scaler)
df_1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(df_1[1], bins=20)
```

```python
plt.hist(df_1[2], bins=20)
```

```python
plt.hist(df_1[3], bins=20)
```

#### 2. Min Max Scaling - CNN- DL
Scales the value between Zero to One
X_scaled = (X - X_min/ X_max- X_min)

```python
from sklearn.preprocessing import MinMaxScaler
minMaxScaler = MinMaxScaler()
df_minmax = pd.DataFrame(minMaxScaler.fit_transform(df),columns=df.columns)
df_minmax
```

```python
plt.hist(df_minmax['Age'], bins=20)
```

```python
plt.hist(df_minmax['Pclass'], bins=20)
```

```python
plt.hist(df_minmax['Fare'], bins=20)
```

#### 3. Roburst Scaler
It is used to scale the feature from median to quantile.

X_scaled = (X- X_median)/IQR, IQR = 75th Quantile - 25th Quantile

```python
from sklearn.preprocessing import RobustScaler
robust_scaler = RobustScaler()
df_robust_scaler = pd.DataFrame(robust_scaler.fit_transform(df),columns=df.columns)
df_robust_scaler
```

```python
plt.hist(df_robust_scaler['Pclass'], bins=20)
```

```python
plt.hist(df_robust_scaler['Age'], bins=20)
```

```python
plt.hist(df_robust_scaler['Fare'], bins=20)
```

### Gaussian Transformation

- Logarithmic Transformation
- Reciprocal Transformation
- Square Root Transformation
- Exponential Transformation
- BoxCox Transformation

```python
df = pd.read_csv('titanic_train.csv', usecols=['Age','Fare','Survived'])
df.tail()
```

```python
plt.figure(figsize=(20,10))
df.hist(bins=20)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df.isna().sum()
```

```python
import scipy.stats as stat
```

```python
def plot_data(df, feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature], dist='norm', plot=plt)
```

```python
plot_data(df,'Age')
```

### Logarithmic Transformation

```python
df['Age_log'] = np.log(df['Age'])
plot_data(df, 'Age_log')
```

```python
plot_data(df, 'Fare')
```

```python
df['Age_reciprocal'] = 1/df.Age
plot_data(df, 'Age_reciprocal')
```

```python
df['Age_sq_root'] = np.sqrt(df.Age)
plot_data(df, 'Age_sq_root')
```

```python
df['Age_exponential'] = df.Age**(1/1.2)
plot_data(df,'Age_exponential')
```

### BoxCox transformation

```python
df['Age_boxcox'], parameters = stat.boxcox(df['Age'])
```

```python
plot_data(df,'Age_boxcox')
```

```python
plot_data(df,'Fare')
```

```python
df["Fare_log"] = np.log1p(df['Fare'])
plot_data(df,'Fare_log')
```
